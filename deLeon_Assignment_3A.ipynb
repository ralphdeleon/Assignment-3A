````python
## Code goes here
import re

import csv    
````

````python
recweek_data = list()

with open('recweekdata.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    count_line = 0
    for row in csv_reader:
        if count_line == 0:
            pass
        else:
            recweek_data.append({"firstname":row[0],"lastname":row[1],"birthmonth":row[2],"birthday":row[3],"birthyear":row[4],"course":row[5],"phonenumber":row[6]})
        count_line += 1
        
recweek_data
````

````python
# birthmonth
dirty_birthmonth = [] #unfiltered data for birthmonth
for i in recweek_data:
    dirty_birthmonth.append(str(i["birthmonth"]))

clean_birth_month = [re.sub(r'(January|Jan)',r'1',month) for month in clean_birth_month]
clean_birth_month = [re.sub(r'(February|Feb)',r'2',month) for month in clean_birth_month]
clean_birth_month = [re.sub(r'(March|Mar)',r'3',month) for month in clean_birth_month]
clean_birth_month = [re.sub(r'(April|Apr)',r'4',month) for month in clean_birth_month]
clean_birth_month = [re.sub(r'(May)',r'5',month) for month in clean_birth_month]
clean_birth_month = [re.sub(r'(June|Jun)',r'6',month) for month in clean_birth_month]
clean_birth_month = [re.sub(r'(July|Jul)',r'7',month) for month in clean_birth_month]
clean_birth_month = [re.sub(r'(August|Aug)',r'8',month) for month in clean_birth_month]
clean_birth_month = [re.sub(r'(September|Sep)',r'9',month) for month in clean_birth_month]
clean_birth_month = [re.sub(r'(October|Oct)',r'10',month) for month in clean_birth_month]
clean_birth_month = [re.sub(r'(November|Nov)',r'11',month) for month in clean_birth_month]
clean_birth_month = [re.sub(r'(December|Dec)',r'12',month) for month in clean_birth_month]

clean_birth_month
````

````python
#birthday
clean_birthday = list()
for i in recweek_data:
    clean_birthday.append(str(i["birthday"]))
````

````python
clean_lastname = list()
for i in recweek_data:
    clean_lastname.append(str(i["lastname"]))
````

````python
clean_firstname = list()
for i in recweek_data:
    clean_firstname.append(str(i["firstname"]))
````

````python
# birthyear
dirty_birthyear = [] #unfiltered data for birthyear
for i in recweek_data:
    dirty_birthyear.append(str(i["birthyear"]))
    

clean_birth_year = [re.sub(r'(1997|97)',r'97',yr) for yr in clean_birth_year]
clean_birth_year = [re.sub(r'(1998|98)',r'98',yr) for yr in clean_birth_year]
clean_birth_year = [re.sub(r'(1999|99)',r'99',yr) for yr in clean_birth_year]
clean_birth_year = [re.sub(r'(2000|00)',r'00',yr) for yr in clean_birth_year]
clean_birth_year = [re.sub(r'(2001|01)',r'01',yr) for yr in clean_birth_year]

clean_birth_year
````

````python
# course
dirty_course = [] #unfiltered data for course
for i in recweek_data:
    dirty_course.append(str(i["course"]))

clean_course = [re.sub(r'(Management|BS MGT|MGT)',r'MGT',course) for course in dirty_course]
clean_course = [re.sub(r'(Information Technology Entrepreneurship|BS ITE|ITE)',r'ITE',course) for course in clean_course]
clean_course = [re.sub(r'(Management Engineering|MGT Engineering|BS ME|ME)',r'ME',course) for course in clean_course]
clean_course = [re.sub(r'(Legal Management|Legal MGT|BS LM|LM)',r'LM',course) for course in clean_course]
clean_course = [re.sub(r'(Communications Technology|BS CTM|CTM)',r'CTM',course) for course in clean_course]

clean_course
````

````python
# phone number
dirty_number = [] #unfiltered data for phone number
for i in recweek_data:
    dirty_number.append(str(i["phonenumber"]))

clean_number = [re.sub(r'\W+63 ',r'',number,1) for number in dirty_number]
clean_number = [re.sub(r'9', r'09',number,1) for number in clean_number]

clean_number
````

````python
clean_data = list()

for i in range (len(clean_number)):
    clean_data.append({"firstname":clean_firstname[i],"lastname":clean_lastname[i],"birthmonth":clean_birth_month[i],"birthday":clean_birthday[i],"birthyear":clean_birth_year[i],"course":clean_course[i],"phonenumber":clean_number[i]})
       
print(clean_data)
````

````python
with open('clean_recweek_data.csv', 'w+', newline='') as csvfile:
    fieldnames = ['firstname','lastname','birthmonth','birthday','birthyear','course','phonenumber']
    writer = csv.DictWriter(csvfile, fieldnames =fieldnames)
    writer.writeheader()
    for i in range(len(clean_firstname)):
        writer.writerow({'firstname':clean_data[i]['firstname'],
                         'lastname':clean_data[i]['lastname'],
                         'birthmonth':clean_data[i]['birthmonth'],
                         'birthday':clean_data[i]['birthday'],
                         'birthyear':clean_data[i]['birthyear'],
                         'course':clean_data[i]['course'],
                         'phonenumber':clean_data[i]['phonenumber']})
````
